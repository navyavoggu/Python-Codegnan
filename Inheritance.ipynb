#INHERITANCE : A class(child can inherite the properties of another class(parent)

```python
#SINGLE INHERITANCE
#parent class
class Animal:
    def legs(self):
        return "Animal has four legs"
#child class
class Dog(Animal):
    def sound(self): #instance method
        return "Dog can make sound like boow boow"
a = Animal()
d = Dog()
print(d.sound())
print(d.legs())
print(a.legs())
```
