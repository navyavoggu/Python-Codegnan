***Regular Expression***

```python
import re
string = "The sky is blue and ocean is blue"
pat = "The"
res = re.match(pat,string)
print(res)
print(res.group()) #it return the matched pattern
print(res.span()) #it returns the matched pattern position
```

```python
import re
string = "The sky is blue and ocean is blue"
pat = "the"
res = re.match(pat,string)
print(res)
print(res.group()) #it return the matched pattern
print(res.span()) #it returns the matched pattern position
```

```python
#search(pat,string):This function check entire string and it returns the first matched pattern
import re
string = "The sky is blue and ocean is blue"
pat = "blue"
res = re.search(pat,string)
print(res)
print(res.group()) #it return the matched pattern
print(res.span()) #it returns the matched pattern position
```

```python
#findall(pat,string):returns all matched patterns in the string
import re
string = "The sky is blue and ocean is blue"
pat = "blue"
res = re.findall(pat,string)
print(res)
```

```python
#findall(pat,string):returns all matched patterns in the string
import re
string = "The sky is blue and ocean is blue"
pat = "[abcd]"
res = re.findall(pat,string)
print(res)
```

```python
#findall(pat,string):returns all matched patterns in the string
import re
string = "The sky2 is bl1ue and oce6an is blu5e"
pat = "[0-9]"
res = re.findall(pat,string)
print(res)
```

```python
#findall(pat,string):returns all matched patterns in the string
import re
string = "The sky2 is bl1ue and oce6an is blu5e"
pat = "[0-9]"
res = re.sub(pat,"",string)
print(res)
```
