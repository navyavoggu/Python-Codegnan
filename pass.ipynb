```python
#pass by reference
def fun(l,val):
   l.append(val)
   return 1
lst=[1,2,3]
val =10
print(fun(lst,val))
print(lst)
```

```python
# pass by value
def fun(x):
   x=x+1
   return x
val =10
print(fun(val))
print(val)
```

```python
#recurssion
def fun():
    print('function')
    fun()
fun()
```

```python
def fun(x):
    if x==0:  #-base statement
      return x #-stopping condition
    print(x)
    fun(x-1)
fun(4)
```

```python
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n+1)
print(fact(10))
```

```python
def fun(n):
    if n==1:
        return
    print(n)
    fun(n-1)
fun(5)
```

```python
#print 1 to n numbers
def fun(n):
    if n==0:
        return
    print(n)
    fun(n-1)
fun(5)
```

```python
#print n to 1 numbers
def fun(n):
    if n==0:
        return
    fun(n-1)
    print(n)
fun()
```

```python
def even_numbers(n):
    if n==0:
        return
    even_numbers(n-1)
    print(n)
    if n%2==0:
        print(n,end=' ')
n=10

even_numbers(n)
```

```python
#print odd numbers between 1 to n
def fun(n):
    if n==0:
        return
    fun(n-1)
    if n%2==1:
       print(n)
fun(5)
```

```python
#print even numbers between 1 to n
def fun(n):
    if n==0:
        return
    fun(n-1)
    if n%2==0:
       print(n)
fun(5)
```
