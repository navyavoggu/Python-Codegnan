```python
#pass by reference
def fun(l,val):
   l.append(val)
   return 1
lst=[1,2,3]
val =10
print(fun(lst,val))
print(lst)
```

```python
# pass by value
def fun(x):
   x=x+1
   return x
val =10
print(fun(val))
print(val)
```

```python
#recurssion
def fun():
    print('function')
    fun()
fun()
```

```python
def fun(x):
    if x==0:  #-base statement
      return x #-stopping condition
    print(x)
    fun(x-1)
fun(4)
```
