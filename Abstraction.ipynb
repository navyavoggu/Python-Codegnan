
***ABSTRACTION***

```python

from abc import ABC, abstractmethod
class Car(ABC):
    #abstract methods
    @abstractmethod
    def engine(self):# method declaration and without implementation
        pass
    @abstractmethod
    def tyres(self):
        pass
    def brakes(self):
        pass
#child class
class honda(Car):
    def engine(self):
        return f"This is honda engine"
    def tyres(self):
        return f"Honda car has 4- bit tyres"
    def brakes(self):
        return "Honda car applies brakes automatically"
    def ac(self):
        return f"Honda car has centeral Ac"
h = honda()
print(h.tyres())
print(h.ac())
    
```
