```python
f = open('test1.txt','r')
text = f.read()# reads entire context as single string
print("---",text[3])# returns 3rd position character
print("****",text[:30])#return first 30 characters
print("-=-=-=-=-",text)
f.seek(0) # tacking cursor to 0th position
print("-----------------------------")
print(f.readline(5))
print(f.readlines())
```

***Directory Operations***

```python
import os
#check curent working directory
print(os.getcwd())
```

```python
# to create new directory 
os.mkdir("sample")
```

```python
#to change the on directory to another directory bys using chdir()
os.chdir("sample")
print(os.getcwd())
```

```python
#to change the current working directory to python
os.chdir("..")
print(os.getcwd())
```

```python
#returns the all fiels and directories in current working directory
print(os.listdir('.'))
```

```python
#create nested folders 
os.makedirs('a/b/c')
```

```python
# delete empty directory 
os.rmdir("sample")
```

```python
#rename(old_name,new_name) - It is used to rename the directory
os.rename("a","A")
```

```python
#check weather the path is exists or not
print(os.path.exists("A"))
```

```python
#check weather the path is exists or not
print(os.path.exists("A/b"))
print(os.path.exists("A/c"))
```
